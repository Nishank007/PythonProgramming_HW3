# Question 1(25 points) :

<B>A  string  is  an  anagram  of  another  if  the  second  string  is  simply  a 
scrambled version of the first. Write a python program to implement the following 
game:<br>
    
a) Reads  in  a  file  that  has  words  and  their  meanings  in  a  text  file.  An  example 
“words  and  their  meanings”  file  is  given  in  Canvas  (i.e.,  mywords.txt).  Note 
that your program needs to ask the user for the “words and their meanings” file 
to use. <br>
b) The words and their meanings text file is of the csv (comma separated values) 
format. Use either notepad++ or notepad to create your file in the same format 
as the mywords.txt file given to you in Canvas.<br>
c) Your  program  should  then  pick  a  word  from  the  “words  and  their  meanings” 
file, scramble the letters, and ask the user to unscramble it. Every run of your 
program should pick a word at random.<br>
d) The user may type in the unscrambled word or may ask for the 
definition/meaning  of  the  word  by  entering  a  question  mark.  The  game 
continues until the user says “no” to the question: “Do you want to continue?”.<br>
Rock-paper-scissors is a surprisingly popular game that many people play 
seriously (see the Wikipedia article for details).<br>   
</B>.
<br>

#### -> Ask the user to enter the filename with words and their meanings. If file doesn't exist, show error.

#### -> Open the file as read-only and split the words and their meanings in a dictionary.

#### -> Shuffle/scramble the words and show it to the user with a help option which the user can select by entering "?".

#### -> If the user enters "?", show the meaning of the scrambled word.

#### -> If the user guesses incorrect, print Wrong! try again.

#### -> If the user guesses correct, ask him if he wants to continue or not. If not, print Goodbye!.

```python
import random

words = {} ## Initialize an empty dictionary

def shuffled(word): ## Shuffling words
    word = list(word)
    random.shuffle(word)
    return ''.join(word)

while True:
    try:
        filename = input('Give the name of the “words and their meanings” file: ') ## Input filename
        with open(filename, 'r') as f:
            for line in f.readlines():
                word, meaning = line.strip().split(",") ## Splitting words as "," as the delimiter
                words[word] = meaning
        break
    except FileNotFoundError:
        print("Invalid Filename")  ## If file doesn't exists

word, meaning = random.choice(list(words.items()))
print('Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word:', shuffled(word))
while True:
    answer = input("Enter the answer [or ? for the meaning]: ") 
    if answer == word: ## If user enters right, ask if he wants to continue
        next = input("You got it! Do you want to continue [yes or no]: ")
        if next == "no":
            break
        else:
            word, meaning = random.choice(list(words.items()))
            print('Unscramble the following letters to form a word. Type “?” for the meaning of'
                  'the unscrambled word:', shuffled(word)) ## If user wants to continue playing
            continue
    if answer == '?': ## Give user the meaning of the scrambled word
        print("The word means:", meaning)
    else:
        print("Wrong, try again") ## If user enters incorrect answer
print("Goodbye!") # Exit
```

# Question 2(25 points) :

<B>The file restaurant_reviews.txt is given to you. <br>
 
The file contains names of reviewers and their reviews of different restaurants in 
the comma separated format. The first string is the name of the reviewer followed 
by a name of a restaurant and its rating. You are required to write a Python program 
that  computes  a  similarity  score  between  any  two  reviewers  using  Euclidean 
distances. An example data from the restaurant_reviews.txt file is given below: <br><br>

    
<i>Pete Wellsworth’s reviews are:  </i><br>
'Rooster and Owl':  3.0, <br> 
'El Sapo Cuban Social Club': 4.0, <br>
 'Punjab Grill': 3.0,  <br>
'Shibumi': 5.0,  <br>
'Mama Chang': 3.5 <br>
    
<i>Jay Samuel’s reviews are:  </i><br>
'El Sapo Cuban Social Club':4.5, <br>
'Mama Chang':1.0, <br>
'Shibumi':4.0 <br>
 
The Euclidean distance between Jay Samuel and Pete Wellsworth is computed as 
follows: <br>

1. Only consider the restaurants that both have reviewed.<br>
2. Take the difference between corresponding reviews. <br>
3. Sum the square of the differences.<br>
4. The square root of the sum of differences is the Euclidean score. The shorter the 
distance the closer the two reviewers. <br>
    
For our example, this would be: (4.0 – 4.5)2  +  (3.5 – 1.0)2 + (5.0 – 4.0)2 = 0.25 + 
6.25 + 1 = 7.5 <br>
Euclidean distance is Square Root of 7.5 = 2.7386 <br>
 
Your program should provide the following: <br>
1. Ability to read in a user provided filename which contains restaurant reviews in 
the comma separated format similar to the given file restaurant_reviews.txt <br>
2. Functionality to compute the similarity between two reviewers provided by the 
user <br>
3. Functionality to compute the similarity between one user provided reviewer and 
all other reviewers in the database <br>
</B>
<br>

#### -> Ask user to enter the restaurant review filename. Check if it exists or give an error.

#### -> Split the text in the file with "," as the delimiter & store the info in a dictionary.

#### -> Ask the user for its choice between similarity between 2 reviewers OR one against all OR exit the program. According to the users choice, move forward.

#### -> If 1, Take the difference between the reviews which both have in common. The square root of the sum of differences is the Euclidean score.

#### -> If 2, Do the above step with all the reviewers in the file.

#### -> If 3, Exit the program.

```python
import math

all_reviews = {}

def similarity(reviewer1, reviewer2):  ## Calculates the similarity
    if reviewer1 in all_reviews.keys():
        if reviewer2 in all_reviews.keys():
            score = 0
            for k in all_reviews[reviewer1].keys():
                if k in all_reviews[reviewer2].keys():
                    score = score + (all_reviews[reviewer1].get(k) - all_reviews[reviewer2].get(k)) * (
                            all_reviews[reviewer1].get(k) - all_reviews[reviewer2].get(k))
            return math.sqrt(score)
        return -2
    return -1

while True:
    try:
        filename = input("Give the name of the restaurant reviews file: ")  ## Inputs filename
        with open(filename, 'r') as f: ## Opens the filename in Read mode
            for line in f: 
                tokens = line.split(',')
                name = tokens[0]
                review = {}

                for i in range(1, len(tokens), 2):
                    restaurant_name = tokens[i]
                    rating = float(tokens[i + 1])
                    review[restaurant_name] = rating
                    all_reviews[name] = review
        break
    except FileNotFoundError:
        print("Invalid Filename")  ## If file doesn't exists

while True:
    try:
        choice = int(input("What  do  you  want  to  do?  Input  1  for  similarity  between  two  reviewers, or "
                           "Input 2 for similarity between one reviewer and all others in the database or 3 to quit:  "))
        if choice == 1:  ## If user chooses 1
            reviewer1 = input("Provide Reviewer1 name: ")  ## Input reviewer1's name
            if not reviewer1:
                print("Invalid Input") 
            reviewer2 = input("Provide Reviewer2 name: ")  ## Input reviewer2's name
            if not reviewer2:
                print("Invalid Input") 

            score = similarity(reviewer1, reviewer2)
            if score == -1:
                print("Reviewer doesn't exist") ## If reviewer1 doesn't exists
            elif score == -2:
                print("Reviewer doesn't exist") ## If reviewer2 doesn't exists
            else:
                print("The similarity score between ", reviewer1, " and ", reviewer2, "is: ", round(score, 4))
        elif choice == 2:  ### If user chooses 2
            reviewer1 = input("Provide Reviewer name: ")  ## Input reviewer1's name
            if reviewer1 in all_reviews.keys():
                print("The Similarity Scores are: ")
                for reviewer2 in all_reviews.keys():
                    if reviewer1 == reviewer2:
                        continue
                    score = similarity(reviewer1, reviewer2)
                    print(reviewer1, "\t", reviewer2, "\t", format(round(score, 2)))
            else:
                print("Reviewer doesn't exist")
        elif choice == 3:  ## If user chooses 3
            print("Goodbye!")
            break
        else:  ### If user enters anything other than 1, 2 or 3
            print("Please enter 1, 2 or 3")
    except ValueError:
        print("Invalid Option")
```
